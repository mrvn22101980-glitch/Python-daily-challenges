```python
#Tuple
product_categories = ("Electronics", "Stationery", "Clothing")
```

```python
# list
products = ["Laptop", "Pen", "T-shirt"]
```

```python
new_product = input("Enter a new product to add: ")
products.append(new_product)
print("Product added!")
```

```python
print("\nCurrent product list with indexes:")
for i, item in enumerate(products):
    print(i, ":", item)
    remove_index = int(input("Enter the index of the product to remove: "))
```

```python
if 0 <= remove_index < len(products):
    removed_item = products.pop(remove_index)
    print(f"{removed_item} removed from the inventory.")
else:
    print("Invalid index! No item removed.")

```

```python
print("\nFinal Product List:", products)
print("Product Categories:", product_categories)

```
