```python
transaction_counter = 0
class BankAccount:

    def __init__(self, account_holder, balance=0):
        self.account_holder = account_holder   # Attribute 1
        self.balance = balance                 # Attribute 2

    def deposit(self, amount):
        global transaction_counter
        if amount > 0:
            self.balance += amount
            transaction_counter += 1
            print(f"Deposited ₹{amount} into {self.account_holder}'s account.")
        else:
            print("Deposit amount must be positive!")

    def withdraw(self, amount):
        global transaction_counter
        # Local variable example
        allowable = self.balance - amount

        if amount <= 0:
            print("Withdrawal amount must be positive!")
        elif allowable < 0:
            print("Withdrawal denied! Insufficient balance.")
        else:
            self.balance -= amount
            transaction_counter += 1
            print(f"Withdrew ₹{amount} from {self.account_holder}'s account.")

    def display_balance(self):
        print(f"{self.account_holder}'s current balance: ₹{self.balance}")


# Creating object using positional arguments
acc1 = BankAccount("revathy", 50000)

# Creating object using keyword arguments
acc2 = BankAccount(account_holder="vaishu", balance=100000)

# Performing transactions
acc1.deposit(2500)          # positional
acc1.withdraw(amount=1000)  # keyword argument
acc1.display_balance()

acc2.deposit(amount=5000)   # keyword
acc2.withdraw(25000)        # positional
acc2.display_balance()

# Display global transaction count
print(f"\nTotal transactions completed: {transaction_counter}")
```
