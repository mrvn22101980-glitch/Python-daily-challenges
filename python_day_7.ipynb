```python
import random

# 1. Randomly choose a number between 1 and 20
secret_number = random.randint(1, 20)
print(" Welcome to the Number Guessing Game!")

Secret_Namuber = 15
attempts = 0
max_attempts = 5
```

```python

# 2. Allow the user to guess up to 5 times
while attempts < max_attempts:
    try:
        guess = int(input(f"\nAttempt {attempts + 1}/5 ‚Äì Enter your guess (1‚Äì20): "))
    except ValueError:
        print("Invalid input! Please enter a valid integer.")
        continue

      # Do NOT count invalid input as an attempt

    # Valid guess ‚Üí count the attempt
    attempts += 1

    # 3. Check guess
    if guess < secret_number:
        print("Too low! Try again.")
    elif guess > secret_number:
        print("Too high! Try again.")
    else:
        # 4. Correct guess ‚Üí break the loop
        print(f"\nüéâ Correct! The number was {secret_number}. You guessed it in {attempts} attempts!")
        break
else:
    # 5. Else executes ONLY if loop ended WITHOUT break
    print(f"\n‚ùå You've used all 5 attempts! The correct number was {secret_number}.")
```
