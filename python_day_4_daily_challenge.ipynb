```python
contacts = [
    ["Revathy", "9876543222", "revathy@example.com"],
    ["madhu",   "8976546780", "madhu@example.com"]]
```

```python
print("Add a new contact:")
name = input("Enter name: ")
phone = input("Enter phone number: ")
email = input("Enter email: ")
contacts.append([name, phone, email])
print("\nContact added successfully!\n")
```

```python
contacts.sort(key=lambda x: x[0])   # sort by the first element (name)

print("All Contacts (Sorted Alphabetically):")
for c in contacts:
    print(f"Name: {c[0]}, Phone: {c[1]}, Email: {c[2]}")
```

```python
search = input("\nEnter name to search (partial match supported): ").lower()

print("\nSearch Results:")
found = False
for c in contacts:
    if search in c[0].lower():  # partial match check
        print(f"Name: {c[0]}, Phone: {c[1]}, Email: {c[2]}")
        found = True

if not found:
    print("No matching contacts found.")
```

```python
print("\nFirst 3 contacts after sorting:")
for c in contacts[:3]:   # slicing to get first 3
    print(f"Name: {c[0]}, Phone: {c[1]}, Email: {c[2]}")
```
